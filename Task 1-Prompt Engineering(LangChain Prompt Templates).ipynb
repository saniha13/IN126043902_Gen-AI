```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

```python
# Task 1: Simple PromptTemplate

template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Example
output_1 = template_1.format(topic="Artificial Intelligence")
print(output_1)
```

```python
# Task 2: Multiple inputs

multi_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test cases
print(multi_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

```python
# Task 3: Different prompt styles

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

```python
# Task 4: Chat-based system

def get_chat_prompt(role, topic):

    if role == "teacher":
        system_message = "You are a helpful teacher."
    elif role == "interviewer":
        system_message = "You are an interviewer asking challenging questions."
    elif role == "motivator":
        system_message = "You are a motivational speaker."
    else:
        system_message = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Talk about {topic}")
    ])

    return chat_template.format_messages(topic=topic)

# Example
messages = get_chat_prompt("teacher", "Neural Networks")

for msg in messages:
    print(msg)
```

```python
# Task 5: Validation function

def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Defaulting to 'formal'")
        tone = "formal"

    return audience, tone

# Example
print(validate_inputs("kids", "fun"))
```

```python
# Task 6: Main prompt generator

def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template_text = "Explain {topic} for {audience} in a {tone} teaching style"
    elif style == "interview":
        template_text = "Ask interview questions on {topic} for {audience} in a {tone} tone"
    elif style == "storytelling":
        template_text = "Explain {topic} for {audience} in a {tone} storytelling style"
    else:
        template_text = "Explain {topic} for {audience} in a {tone} tone"

    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template_text
    )

    return prompt_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
# Task 7: Reusability test

reusable_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

topics = ["AI", "Python", "Data Science", "Cloud Computing", "Cyber Security"]

for t in topics:
    print(reusable_template.format(topic=t))
```
